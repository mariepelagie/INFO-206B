```python
import os
from strip_word import *
import numpy as np
```

```python
def freq_gen(my_dict,filename):
    path = 'C:/Users/mmoud/Documents/INFO 206B/data/'+filename
    with open(path,'r', encoding="utf8") as lines:
        counter = 0
        for line in lines: #iterate on each line of file
            words = line.split(" ")
            for w in words:
                w = strip_word(w)
                if w!='':  ## count number of words that are not empty spaces
                    counter+=1
                ## generate word frequency dictionary
                if w in my_dict:
                    my_dict[w]+=1
                else:
                    my_dict[w] = 1
    return counter # return number of words that excluding empty spaces
                
```

```python
def build_dictionary(prefix, my_dict):
#     my_dict ={} # initialize dictionary
    path = 'C:/Users/mmoud/Documents/INFO 206B/data' ## file path
    #files = [] #initialize list of filenames
    files_prefix = [] ## initialise list of all filenames started with selected prefix
    wordcount = 0
    for i in os.listdir(path):
        #files.append(i)
        if os.path.isfile(os.path.join(path,i)) and prefix in i:
            files_prefix.append(i)
    if prefix == "":
        for f in files_prefix:
            wordcount+=freq_gen(my_dict,f)
    if prefix[:2] == "pk" or prefix[:2] == "dr":
        if len(prefix) == 2:
            for f in files_prefix:
                wordcount+=freq_gen(my_dict,f)
        elif len(prefix) > 2:
            for f in files_prefix:
                if f[:-4] ==prefix:
                    wordcount+=freq_gen(my_dict,f)
                    
    return wordcount

           
```
